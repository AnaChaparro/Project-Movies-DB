## 4. Tiramos Queries para comprobar si el funcionamiento de la base de datos es correcto.

#### ¿Cuáles son los 5 actores con más películas de duración superior a 80 minutos?

SELECT a.full_name, count(o.film_id) AS film_qty, group_concat(f.title)
FROM actor AS a

LEFT JOIN old_hdd AS o
ON a.actor_id = o.actor_id

LEFT JOIN film AS f
ON f.film_id = o.film_id 

WHERE f.length > 80

GROUP BY full_name
ORDER BY film_qty DESC
LIMIT 5;

#### ¿Cuáles son los 10 películas de la categoría Classics con contengan un Behind the Scenes?

SELECT f.title
FROM film AS f

LEFT JOIN old_hdd AS o
ON f.film_id = o.film_id

LEFT JOIN category AS c
ON c.category_id= o.category_id

WHERE c.name= 'Classics' AND  f.behind_the_scenes= 'Yes'

LIMIT 10;

#### Los actores con más películas en inglés de la categoría Children.

SELECT a.full_name, f.title
FROM actor AS a

LEFT JOIN old_hdd AS o
ON a.actor_id = o.actor_id

LEFT JOIN film AS f
ON f.film_id = o.film_id

LEFT JOIN category AS c
ON c.category_id= o.category_id

LEFT JOIN language AS l
ON l.language_id= f.language_id

WHERE l.name = 'English' and c.name = 'Children';

#### ¿Cuales son las 15 películas con mayor número de actores que intervienen?

SELECT f.title, count(a.full_name) AS number_of_actors
FROM film as f

LEFT JOIN old_hdd as o
ON o.film_id = f.film_id

LEFT JOIN actor as a
ON a.actor_id=o.actor_id


GROUP BY f.title
ORDER BY number_of_actors DESC
LIMIT 15;

#### ¿Cuáles son las películas de acción de una duración inferior a 100 minutos?


SELECT f.title, f.length
FROM film AS f

LEFT JOIN old_hdd AS o
ON f.film_id = o.film_id

LEFT JOIN category AS c
ON c.category_id = c.category_id

WHERE c.name = 'Action' AND f.length <100

GROUP BY f.film_id;

#### Actores que han salido en películas con clasificación 'G'.

SELECT a.full_name, f.title
FROM actor AS a

LEFT JOIN old_hdd AS o
ON a.actor_id = o.actor_id

LEFT JOIN film AS f
ON f.film_id = o.film_id

#### Las 20 películas en la que solo aparece un actor con menor duración? 

SELECT f.title AS "TITLTE", a.full_name AS "NAME", f.length AS "DURATION"
FROM film AS f

LEFT JOIN old_hdd AS o
ON o.film_id = f.film_id

LEFT JOIN actor AS a
ON a.actor_id = o.actor_id

WHERE a.actor_id = 1 

GROUP BY f.film_id
ORDER BY f.length ASC
LIMIT 20;

#### Determinar cuántas películas existen por cada categoría.

SELECT c.name AS "CATEGORY_NAME", count(c.category_id) AS "NUMBER_OF_FILMS"
FROM category AS c

LEFT JOIN old_hdd AS o
ON o.category_id = c.category_id

LEFT JOIN film AS f
ON f.film_id = o.film_id


GROUP BY c.name
ORDER BY NUMBER_OF_FILMS ASC;
